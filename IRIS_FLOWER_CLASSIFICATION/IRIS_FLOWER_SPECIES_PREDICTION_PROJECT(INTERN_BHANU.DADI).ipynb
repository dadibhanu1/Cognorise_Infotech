## HYPERPARAMETER TUNING


### GRIDSEARCH CV

```python
max_depth = np.linspace(3, 10, 6,dtype="int")
min_samples_split = [2, 5, 10, 15, 30, 35]
min_samples_leaf = [1, 2, 5, 10, 15]
criterion = ["gini","entropy"]
min_impurity_decrease = [0.0001, 0.0005,0.1]
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc_model=DecisionTreeClassifier()
```

```python
param_grid={
    "max_depth" :max_depth ,
    "min_samples_split" :min_samples_split,
    "min_samples_leaf":min_samples_leaf,
    "criterion":criterion,
    "min_impurity_decrease" : min_impurity_decrease
}
```

```python
gsv=GridSearchCV(dtc_model,param_grid,cv=5,n_jobs=-1,verbose=10)
```

```python
gsv
```

###  HYPERPARAMETER TUNING

### GRIDSEARCH CV


```python
n_estimators=[3,4,5,6,7,8,9,10]
criterion=['gini','entropy']
max_depth = np.linspace(3, 10, 6,dtype="int")#linspace--means diving the no.s into equal parts
min_samples_split = [2, 5, 10, 15, 30, 35]
min_samples_leaf = [1, 2, 5, 10, 15]
min_impurity_decrease = np.linspace(0.0001, 0.0005, 8)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={"n_estimators":n_estimators,"criterion":criterion,"max_depth" :max_depth ,"min_samples_split" :min_samples_split,
            "min_samples_leaf":min_samples_leaf,
          "min_impurity_decrease" : min_impurity_decrease}
```

```python
gsv=GridSearchCV(rfc_model,param_grid=param_grid,cv=3,n_jobs=-1,verbose=10)
```

### BAR PLOT- ACCURACY SCORES OF TRAIN & TEST DATA  OF ALL THE MODEL'S

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Algorithms=["lor_model","knn_model","svc_model","dtc_model","rfc_model"]
Accuracy_train_scores=[0.973,0.991,0.991,0.982,0.964]
Accuracy_test_scores=[0.947,0.921,0.973,0.947,0.921]
```

```python
x = np.arange(len(Algorithms))
bar_width=0.35
width = 0.35
```

```python
plt.figure(figsize=(10,6))
plt.bar(x - bar_width/2, Accuracy_train_scores, bar_width, label='Accuracy_train_scores', color='#ADD8E6',alpha=0.7)
plt.bar(x + bar_width/2,Accuracy_test_scores, bar_width, label=' Accuracy_test_scores', color='#3366cc', alpha=0.5)


plt.xlabel("Algorithms")
plt.ylabel("Accuracy Scores")
plt.title('Accuracy Scores of Different Algorithms for Flower species prediction')
plt.xticks(x,Algorithms)
plt.legend(loc='lower left')
#plt.show()


for i in range(len(Algorithms)):
    plt.annotate(Accuracy_train_scores[i],(x[i] - width/2,Accuracy_train_scores[i]),ha='center',  fontsize=10)
    plt.annotate(f'{Accuracy_test_scores[i]}', (x[i] + width/2,Accuracy_test_scores[i]), ha='center', fontsize=10)
plt.show()
```
